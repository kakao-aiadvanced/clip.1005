```python
from typing import List
import bs4
from langchain import hub
from langchain_chroma import Chroma
from langchain_community.document_loaders import WebBaseLoader
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.documents import Document


def create_chroma_from_urls(urls: List[str]):
    vectorstore = Chroma(embedding_function=OpenAIEmbeddings())

    # Load, chunk and index the contents of the blog.
    loader = WebBaseLoader(
        web_paths=urls,
        bs_kwargs=dict(
            parse_only=bs4.SoupStrainer(
                class_=("post-content", "post-title", "post-header")
            )
        ),
    )
    docs = loader.load()

    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    vectorstore.add_documents(documents=splits)

    
    return vectorstore

def create_llamaindex_from_urls(urls: List[str]):
    pass

def create_retriever_from_urls(urls: List[str], db_type='chroma'):
    if db_type == "chroma":
        return create_chroma_from_urls(urls).as_retriever()
    elif db_type == "llamaindex":
        pass
```

```python
urls = [
    "https://lilianweng.github.io/posts/2023-06-23-agent/",
    "https://lilianweng.github.io/posts/2023-03-15-prompt-engineering/",
    "https://lilianweng.github.io/posts/2023-10-25-adv-attack-llm/",
]

retriever = create_retriever_from_urls(urls)


prompt = hub.pull("rlm/rag-prompt")

print("prompt : \n")
print(prompt)


def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)


rag_chain1 = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

res = rag_chain1.invoke("agent memory")

print(res)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.output_parsers import ResponseSchema, StructuredOutputParser
from langchain_core.runnables import RunnableLambda, RunnablePassthrough

response_schemas = [
    ResponseSchema(name="relevance", description="Answer to whether the retrieved chunk is relevant to the user query. The value should be 'yes' or 'no'."),
    ResponseSchema(name="reason", description="reason for result"),
]
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)

# llm_with_structure = llm.with_structured_output(output_parser)

print(output_parser.get_format_instructions())

relevance_prompt = ChatPromptTemplate.from_template(
    template="""
        <chunk>
        {context}
        </chunk>

        <user query>
        {question}
        </user query>

        {ins}
    """ #+ output_parser.get_format_instructions()
)

rag_chain2 = (
    {"context": retriever | format_docs, "question": RunnablePassthrough(), "ins": RunnableLambda(lambda x: output_parser.get_format_instructions())}
    | relevance_prompt
    | llm
    | output_parser
)

res = rag_chain2.invoke("agent memory")
print(res)
```

```python
def route(ctx):
    q = ctx['question']
    relevance_res = ctx['relevance_res']
    if relevance_res['relevance'] == "yes":
        return rag_chain1.invoke(q)
    else:
        raise Exception("invalid relevance")


rag_chain3 = { 'relevance_res':rag_chain2, "question": RunnablePassthrough() } | RunnableLambda(route)

res = rag_chain3.invoke("agent memory")
print(res)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain.output_parsers import ResponseSchema, StructuredOutputParser
from langchain_core.runnables import RunnableLambda, RunnablePassthrough

response_schemas = [
    ResponseSchema(name="relevance", description="Answer to whether hallucination chunk is exist to the user query, chunk, response. The value should be 'yes' or 'no'."),
    ResponseSchema(name="reason", description="reason for result"),
]
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)

# llm_with_structure = llm.with_structured_output(output_parser)


hallucination_prompt = ChatPromptTemplate.from_template(
    template="""
        <chunk>
        {context}
        </chunk>

        <user query>
        {question}
        </user query>

        <response>
        {response}
        </response>

        {ins}
    """ 
)

rag_chain4 = (
    {"context": retriever | format_docs, 
     "question": RunnablePassthrough(), 
     "ins": RunnableLambda(lambda x: output_parser.get_format_instructions()),
     "response": rag_chain3
     }
    | hallucination_prompt
    | llm
    | output_parser
)

res = rag_chain4.invoke("agent memory")
print(res)
```
